```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

xs = torch.Tensor(np.arange(-10, 10, 0.1))

relu = nn.ReLU()
ys = relu(xs).numpy()

## set theme
plt.style.use('ggplot')
plt.plot(xs.numpy(), ys)
plt.ylim(-1, 10)
plt.xlim(-10, 10)
plt.grid(alpha=0.8)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title("ReLU")
plt.savefig('/Users/oriol.alas/PROJECTS/PERSONAL/oriolac.github.io/static/posts/2026/act-loss-functions/relu.png', bbox_inches='tight')
plt.show()
```

```python
relu = nn.Sigmoid()
ys = relu(xs).numpy()

## set theme
plt.style.use('ggplot')
plt.plot(xs.numpy(), ys)
plt.ylim(-1.5, 1.5)
plt.xlim(-10, 10)
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(alpha=0.8)
plt.title("Sigmoid")
plt.savefig('/Users/oriol.alas/PROJECTS/PERSONAL/oriolac.github.io/static/posts/2026/act-loss-functions/sigmoid.png', bbox_inches='tight')
plt.show()
```

```python
relu = nn.Sigmoid()
ys = relu(xs).numpy()

## set theme
plt.style.use('ggplot')
plt.plot(xs.numpy(), ys)
plt.ylim(-1.5, 1.5)
plt.xlim(-10, 10)
plt.grid(alpha=0.8)
plt.title("Sigmoid")
plt.savefig('/Users/oriol.alas/PROJECTS/PERSONAL/oriolac.github.io/static/posts/2026/act-loss-functions/sigmoid.png', bbox_inches='tight')
plt.show()
```

```python
class DerivativeSigmoid(nn.Module):

    def __init__(self):
        super().__init__()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(x) * (1 - self.sigmoid(x))

relu = DerivativeSigmoid()
ys = relu(xs).numpy()

## set theme
plt.style.use('ggplot')
plt.plot(xs.numpy(), ys, color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(-10, 10)
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(alpha=0.8)
plt.title("Derivative of Sigmoid")
plt.savefig('/Users/oriol.alas/PROJECTS/PERSONAL/oriolac.github.io/static/posts/2026/act-loss-functions/der_sigmoid.png', bbox_inches='tight')
plt.show()
```

```python
class DerivativeTanh(nn.Module):

    def __init__(self):
        super().__init__()
        self.tanh = nn.Tanh()

    def forward(self, x):
        return 1 - self.tanh(x) ** 2

relu = DerivativeTanh()
ys = relu(xs).numpy()

## set theme
plt.style.use('ggplot')
plt.plot(xs.numpy(), ys, color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(-10, 10)
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(alpha=0.8)
plt.title("Derivative of Tanh")
plt.savefig('/Users/oriol.alas/PROJECTS/PERSONAL/oriolac.github.io/static/posts/2026/act-loss-functions/der_tanh.png', bbox_inches='tight')
plt.show()
```
