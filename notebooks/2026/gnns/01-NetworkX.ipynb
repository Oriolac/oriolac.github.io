```python
from matplotlib import pyplot as plt
import networkx as nx

g = nx.Graph()

g.add_edge(0, 1)

g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(3, 1)
g.add_edge(3, 4)
g.add_edge(4, 1)
g.add_edge(4, 5)
g.add_edge(5, 4)
g.add_edge(5, 3)
g.add_edge(5, 2)
g.add_edge(5, 6)

def plot_g(g):
    fig, ax = plt.subplots(1,1,)

    pos = nx.spring_layout(
        g,
        seed=42,
        k=1.8,
        iterations=300
    )

    nx.draw_networkx_edges(
        g,
        pos,
        ax=ax,
        arrows=True,
        arrowsize=18,
        width=1.2,
        alpha=0.65,
        connectionstyle='arc3,rad=0.08',
        min_source_margin=18,
        min_target_margin=18
    )

    # Draw nodes
    nx.draw_networkx_nodes(
        g,
        pos,
        ax=ax,
        node_size=900,
        linewidths=0,
        edgecolors="black",
        node_color="#65bb7992",
    )

    nx.draw_networkx_labels(
        g,
        pos,
        ax=ax,
        font_size=12,
        font_color="white",
        font_weight="bold",
    )
    plt.axis('off')

plot_g(g)
acc = 0.1
x_ = 0.38
plt.text(x_, 0.26, "Degrees", fontdict={
    "fontsize": 14,
    "fontweight": "bold",
    "family": "sans-serif",
})
for n in g.nodes():
    y = acc - (n) / 7
    t = plt.text(x_, y, f"Degree of node {n}: {nx.degree(g)[n]}", fontdict={
        "fontsize": 14,
        "family": "sans-serif",
    })

diameter = nx.diameter(g)
plt.text(x_, 0.64, "Infromation", fontdict={
    "fontsize": 14,
    "fontweight": "bold",
    "family": "sans-serif",
})
plt.text(x_, 0.48, f"Diameter: {diameter}", fontdict={
    "fontsize": 14,
    "family": "sans-serif",
})
plt.savefig('../../../static/posts/2026/gnn/basic-info.png')
```
